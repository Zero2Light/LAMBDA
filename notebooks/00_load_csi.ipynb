# 00 Load CSI

Inspect one released LAMBDA `csi_*.npz` file and print path-level fields.

```python
from pathlib import Path

CSI_FILE = Path("path/to/csi_000000.npz")
TOP_PATHS = 5

CSI_FILE
```

```python
from lambda_rf.tools.read_csi import load_csi_npz, print_file_summary

if not CSI_FILE.is_file():
    raise FileNotFoundError(f"Update CSI_FILE first: {CSI_FILE}")

summary = load_csi_npz(CSI_FILE)
print_file_summary(summary, top=TOP_PATHS)
```

```python
arrays = summary["arrays"]
print("Available fields:")
for name, value in arrays.items():
    print(f"{name:32s} shape={value.shape} dtype={value.dtype}")
```
