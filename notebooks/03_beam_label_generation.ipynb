# 03 Beam Label Generation

Generate DFT-codebook beam labels from released path-level CSI.

```python
from pathlib import Path
import subprocess
import sys

REPO_ROOT = Path.cwd()
if not (REPO_ROOT / "examples" / "generate_beam_labels.py").is_file():
    REPO_ROOT = REPO_ROOT.parent

CSI_DIR = Path("path/to/csi/f4p9GHz_V")
OUTPUT_CSV = Path("beam_labels.csv")
CARRIER_FREQUENCY_HZ = 4.9e9
LIMIT = 100

CSI_DIR
```

```python
cmd = [
    sys.executable,
    str(REPO_ROOT / "examples" / "generate_beam_labels.py"),
    "--input-dir", str(CSI_DIR),
    "--output-csv", str(OUTPUT_CSV),
    "--carrier-frequency", str(CARRIER_FREQUENCY_HZ),
    "--array-shape", "8,8",
    "--codebook-shape", "16,16",
    "--limit", str(LIMIT),
]
subprocess.run(cmd, check=True)
OUTPUT_CSV
```
