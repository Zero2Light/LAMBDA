# 01 Array and Subcarrier CSI

Generate array/MIMO CSI and OFDM-like subcarrier CSI from released path-level CSI.

```python
from pathlib import Path

CSI_DIR = Path("path/to/csi/f60p0GHz_V")
OUTPUT_ROOT = Path("path/to/derived_csi")
TX_SHAPE = "4,4"
RX_SHAPE = "1,1"
SUBCARRIER_PROFILE = "sub6_30k_1024"
LIMIT = 5

CSI_DIR
```

```python
from lambda_rf.generate_array_csi import run as run_array_csi

array_dir = OUTPUT_ROOT / "array_csi" / CSI_DIR.name / "rx1x1_tx4x4"
run_array_csi(
    input_dir=str(CSI_DIR),
    output_dir=str(array_dir),
    tx_shape=TX_SHAPE,
    rx_shape=RX_SHAPE,
    spacing_wavelengths=0.5,
    skip_existing=True,
    start_frame=None,
    limit=LIMIT,
    tx_orientation_pose=None,
    rx_orientation_pose=None,
)
array_dir
```

```python
from lambda_rf.generate_subcarrier_csi import run as run_subcarrier_csi

subcarrier_dir = OUTPUT_ROOT / "subcarrier_csi" / CSI_DIR.name / "rx1x1_tx4x4" / SUBCARRIER_PROFILE
run_subcarrier_csi(
    input_dir=str(array_dir),
    output_dir=str(subcarrier_dir),
    profile=SUBCARRIER_PROFILE,
    num_subcarriers=None,
    subcarrier_spacing_hz=None,
    input_mode="array",
    tx_shape=TX_SHAPE,
    rx_shape=RX_SHAPE,
    skip_existing=True,
    start_frame=None,
    limit=LIMIT,
)
subcarrier_dir
```
